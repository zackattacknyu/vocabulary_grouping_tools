```python
from vocabulary_grouping_tools.ingestion.api_ingest import get_english_translation
import pandas as pd
import yaml
```

```python
secrets = yaml.safe_load(open('../metadata/secrets.yml', 'r'))
```

```python
vocab = pd.read_csv('../input_files_from_output/vocab_ak_1_2_ttmik_1_2_w_hanja_refined.csv')
```

```python
vocab
```

```python
hangul_words = vocab['hangul_word']
```

```python
hangul_sylls = set(''.join(hangul_words))
```

```python
hangul_sylls
```

```python
len(hangul_sylls)
```

```python
secrets
```

```python
translation_for_hangul_syll = {}
for syll in hangul_sylls:
    print('Now getting translation for ' + syll)
    english_trans = get_english_translation(syll, secrets['client_id'], secrets['client_secret'])
    print(f'Translation is {english_trans}')
    print()
    translation_for_hangul_syll[syll] = english_trans
```

```python
df_of_syllable_translation = pd.DataFrame(translation_for_hangul_syll.items()).rename(columns={
    0: 'hangul_syllable',
    1: 'english_translation'
})
```

```python

df_of_syllable_translation.to_csv('../metadata/hangul_syllables_translation.csv')
```
